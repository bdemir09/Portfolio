```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
#rcParams['figure.figsize'] = 10, 7
rcParams.update({'font.size': 35})
```

```python
rmsd_d1_015 = np.loadtxt('trajrmsd_D1_015.dat')
rmsd_d2_015 = np.loadtxt('trajrmsd_D2_015.dat')
rmsd_d3_015 = np.loadtxt('trajrmsd_D3_015.dat')

rmsd_d1_050 = np.loadtxt('trajrmsd_D1_050.dat')
rmsd_d2_050 = np.loadtxt('trajrmsd_D2_050.dat')
rmsd_d3_050 = np.loadtxt('trajrmsd_D3_050.dat')
```

```python
time = rmsd_d1_015[:, 0] * 0.004
```

```python
# Create Fig and gridspec
fig = plt.figure(figsize=(60, 15), dpi= 300)
grid = plt.GridSpec(4, 4, hspace=0.5, wspace=0.2)

# Define the axes
ax_main = fig.add_subplot(grid[:-1, :-1])
ax_right = fig.add_subplot(grid[:-1, -1], yticklabels=[])

# Lineplot on main ax
ax_main.plot(time, rmsd_d1_015[:,1], '#1E90FF', label='D1 - 0.15M',linewidth=3)
ax_main.plot(time, rmsd_d2_015[:,1], '#FF83FA', label='D2 - 0.15M',linewidth=3)
ax_main.plot(time, rmsd_d3_015[:,1], '#32CD32', label='D3 - 0.15M',linewidth=3)
ax_main.plot(time, rmsd_d1_050[:,1], '#3D59AB', label='D1 - 0.50M',linewidth=3)
ax_main.plot(time, rmsd_d2_050[:,1], '#9370DB', label='D2 - 0.50M',linewidth=3)
ax_main.plot(time, rmsd_d3_050[:,1], '#2E8B57', label='D3 - 0.50M',linewidth=3)

ax_right.hist(rmsd_d1_015[:,1], bins=100, range=[0, 14], histtype='stepfilled', orientation='horizontal', alpha = 0.4, color ='#1E90FF', edgecolor ='#1E90FF', linewidth=3)
ax_right.hist(rmsd_d2_015[:,1], bins=100, range=[0, 14], histtype='stepfilled', orientation='horizontal', alpha = 0.4, color ='#FF83FA', edgecolor ='#FF83FA', linewidth=3)
ax_right.hist(rmsd_d3_015[:,1], bins=100, range=[0, 14], histtype='stepfilled', orientation='horizontal', alpha = 0.4, color ='#32CD32', edgecolor ='#32CD32', linewidth=3)
ax_right.hist(rmsd_d1_050[:,1], bins=100, range=[0, 14], histtype='stepfilled', orientation='horizontal', alpha = 0.4, color ='#3D59AB', edgecolor ='#3D59AB', linewidth=3)
ax_right.hist(rmsd_d2_050[:,1], bins=100, range=[0, 14], histtype='stepfilled', orientation='horizontal', alpha = 0.4, color ='#9370DB', edgecolor ='#9370DB', linewidth=3)
ax_right.hist(rmsd_d3_050[:,1], bins=100, range=[0, 14], histtype='stepfilled', orientation='horizontal', alpha = 0.4, color ='#2E8B57', edgecolor ='#2E8B57', linewidth=3)


ax_main.set_xlim([20, 200])
ax_main.set_ylim([0, 20])

ax_right.set_xlim([0, 5000])
ax_right.set_ylim([0, 20])
ax_right.set_xlabel('Count')
#ax_right.set_ylabel('RMSD ($\AA$)')

ax_main.set_xlabel('Time (ns)')
ax_main.set_ylabel('RMSD ($\AA$)')
ax_main.legend(loc='upper left', ncol=2);

#fig.savefig('RMSD_comparison.png', dpi=1000)
```
