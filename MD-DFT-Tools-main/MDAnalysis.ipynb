```python
R = rms.RMSD(dnaorigami,  # universe to align
             dnaorigami,  # reference universe or atomgroup
             select='nucleic',  # group to superimpose and calculate RMSD
             ref_frame=0)  # frame index of the reference
R.run()
```

```python
R.rmsd.shape
```

```python
R.rmsd
```

```python
df = pd.DataFrame(R.rmsd,
                  columns=['Frame', 'Time (ns)',
                           'RMSD-Nucleic'])

df
```

```python
ax = df.plot(x='Frame', y=['RMSD-Nucleic'],
             kind='line')
ax.set_ylabel(r'RMSD ($\AA$)')
```
